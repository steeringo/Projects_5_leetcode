## Preprocessing

```python
df.drop(['Name', 'Type'], axis=1, inplace= True)
df.head()
```

```python
df['Onroad Price'] = df['Onroad Price'].astype(str)
df['Onroad Price'] = df['Onroad Price'].str.replace(',', '')
df['Onroad Price'] = df['Onroad Price'].astype(int)
df
```

```python
df['Average Price'] = (df['Min Price (Lakh)']+df['Max Price (Lakh)'])/2
df
```

## Analysing data

**Heatmap of correlation**

```python
corr_matrix = df.corr()
plt.figure(figsize=(13,13))
sns.heatmap(corr_matrix, annot=True ,cmap='coolwarm', fmt= '.2f')
#sns.heatmap(corr_matrix, cmap='coolwarm', fmt= '.2f')
plt.show()
```
